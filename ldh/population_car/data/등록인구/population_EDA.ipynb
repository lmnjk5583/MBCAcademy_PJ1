```python
import pandas as pd

# 1. ÏõêÎ≥∏ CSV Î∂àÎü¨Ïò§Í∏∞
df = pd.read_csv("population_wide.csv")

# üî• 2. Ïà´Ïûê Ïª¨ÎüºÏóêÏÑú ',' Ï†úÍ±∞ + int Î≥ÄÌôò
num_cols = [col for col in df.columns if col not in ["district_id", "district"]]
df[num_cols] = (
    df[num_cols]
    .replace(",", "", regex=True)
    .astype(int)
)

rows = []

# 3. wide ‚Üí long Î≥ÄÌôò
for _, row in df.iterrows():
    district_id = row["district_id"]
    district = row["district"]

    for year in range(2020, 2026):
        population = row[f"population_{year}"]

        if year == 2020:
            population_diff = None
        else:
            population_diff = row[f"population_diff_{year}"]

        rows.append({
            "district_id": district_id,
            "district": district,
            "year": year,
            "population": population,
            "population_diff": population_diff
        })

# 4. DataFrame ÏÉùÏÑ±
population_long_df = pd.DataFrame(rows)

# 5. CSV Ï†ÄÏû•
population_long_df.to_csv(
    "population_long.csv",
    index=False,
    encoding="utf-8-sig"
)

print("population_long.csv ÏÉùÏÑ± ÏôÑÎ£å")
print(population_long_df.head())
```
