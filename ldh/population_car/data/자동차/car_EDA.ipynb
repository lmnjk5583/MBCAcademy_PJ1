```python
import pandas as pd

# 1. ì›ë³¸ CSV ë¶ˆëŸ¬ì˜¤ê¸°
df = pd.read_csv("car_wide.csv")

# ğŸ”¥ 2. ìˆ«ì ì»¬ëŸ¼ì—ì„œ ',' ì œê±° + int ë³€í™˜
num_cols = [col for col in df.columns if col not in ["district_id", "district"]]
df[num_cols] = (
    df[num_cols]
    .replace(",", "", regex=True)
    .astype(int)
)

rows = []

# 3. wide â†’ long ë³€í™˜
for _, row in df.iterrows():
    district_id = row["district_id"]
    district = row["district"]

    for year in range(2020, 2026):
        car_count = row[f"car_count_{year}"]

        if year == 2020:
            car_diff = None
        else:
            car_diff = row[f"car_diff_{year}"]

        rows.append({
            "district_id": district_id,
            "district": district,
            "year": year,
            "car_count": car_count,
            "car_diff": car_diff
        })

# 4. DataFrame ìƒì„±
car_long_df = pd.DataFrame(rows)

# 5. CSV ì €ì¥
car_long_df.to_csv(
    "car_long.csv",
    index=False,
    encoding="utf-8-sig"
)

print("car_long.csv ìƒì„± ì™„ë£Œ")
print(car_long_df.head())
```
